```python
# Hacker Rank puzzle :-)
def calc (n, m, track):
    grid = [[1 for _m in range(m)] for _n in range(n)]
    #print(grid)
    print('grid created')
    i=0
    for trk in track:
        i+=1
        if i%10==0:
            print(i, end=' ')
        for loc in range (trk[1]-1, trk[2]):
            #print(trk[0]-1, loc)
            grid[trk[0]-1][loc] = 0
    #print(grid)
    return sum([sum(rw) for rw in grid])
```

```python
def find_free_slots(exmp, mt):
    flattened = set([val for sublist in exmp for val in sublist])
    mn, mx = min(flattened), max(flattened)
    free_before = mn - 1
    free_after = mt - mx
    remaining = [1 for _ in range(mn, mx + 1)]
    #print(remaining)
    for ex in exmp:
        #print(ex, ex[0] - mn, ex[1] - mn + 1)
        for cl in range(ex[0] - mn, ex[1] - mn + 1):
            remaining[cl] = 0
            #print(remaining)
    return sum(remaining) + free_after + free_before
```
